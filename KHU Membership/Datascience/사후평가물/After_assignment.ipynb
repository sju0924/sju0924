# 컬럼 이름 변경

```python
df.columns[100:110]
```

```python
idx = ['총인구수','연령구간인구수']
for i in range(101):
    if i < 100:
        idx.append(str(i)+'세')
    else:
        idx.append(str(i)+'세 이상')
```

```python
df_m = df.iloc[:,0:103]
df_w = df.iloc[:,103:]

df_m.columns = idx
df_w.columns = idx
df_t.columns = idx
```

# 인구가 0인 지역 제거

```python
df_m = df_m[df_m['총인구수']!=0]
df_w = df_w[df_w['총인구수']!=0]
df_t = df_t[df_t['총인구수']!=0]
```

# 연령별 인구비율 구하기

```python
df_m = df_m.div(df_m['총인구수'], axis=0)
del df_m['총인구수'], df_m['연령구간인구수']

df_w = df_w.div(df_w['총인구수'], axis=0)
del df_w['총인구수'], df_w['연령구간인구수']

df_t = df_t.div(df_t['총인구수'], axis=0)
del df_t['총인구수'], df_t['연령구간인구수']

df_w
```

# 관심있는 지역의 인구비율
관심있는 지역이라 하면<br>
고향, 지금 사는 곳, 학교가 있는 곳 정도?

```python
itr = int(input('입력할 값의 갯수를 입력하시오'))
for i in range (itr):
    name = input('관심있는 지역의 이름을 입력해주세요. :')
    if(i != 0):
        loc1 += df_t.index.str.contains(name)
        loc2 += df_m.index.str.contains(name)
        loc3 += df_w.index.str.contains(name)
    else:
        loc1 = df_t.index.str.contains(name)
        loc2 = df_m.index.str.contains(name)
        loc3 = df_w.index.str.contains(name)
    
    




dfT = df_t[loc1]
dfM = df_m[loc2]
dfW = df_w[loc3]

print(dfT)
```

```python
#pyplot.rc('font', family='경기천년제목OTF_Light.otf')
fontprop = fm.FontProperties(fname='C:\\WINDOWS\\Fonts\\경기천년제목_Light.ttf').get_name()
plt.rc('font', family=fontprop)

plt.rcParams['figure.figsize'] = (8,5)


dfT.T.plot()
pyplot.title('전체')
dfW.T.plot()
pyplot.title('남성')
dfM.T.plot()
pyplot.title('여성')
pyplot.show()
```

```python
for i in range(len(dfT)):
    tempdf = pandas.DataFrame(columns = idx)
    del tempdf['총인구수'], tempdf['연령구간인구수']
    tempdf.loc['전체'] = list(dfT.iloc[i])
    tempdf.loc['남성'] = list(dfM.iloc[i])
    tempdf.loc['여성'] = list(dfW.iloc[i])

    tempdf.T.plot()
    pyplot.title(dfM.index[i])
    pyplot.show()
```

경희대 국캠에 공대가 있어서 그런가, 20대에 남성이 많다.<br>
지금 사는 김포는 신도시라 그런가, 유소년층과 젊은 부모들의 비율이 높다.<br>
반면 흑석동은, 여기도 대학가 동네라 그런가 20대가 비교적 많다. 그리고 노년층의 비율이 비교적 많은데, 지금은 재개발되었지만 이 통계를 작성했던 2014년도에는 달동네가 꽤나 있었다. 그리고 꽤 큰 규모의 재래시장도 있었다. 그쪽으로 가면 어르신들을 많이 볼 수 있었던 기억이 떠오르는 듯 하다.

# 관심있는 지역과 가장 비슷한 연령별 인구비율을 보이는 지역을 구하기

## 전체

```python
plt.rcParams['figure.figsize'] = (16,9)
for i in range(itr):
    df_t.loc[np.power(df_t.sub(dfT.iloc[i], axis=1),2).sum(axis=1).sort_values().index[:5]].T.plot()
    pyplot.title(dfT.index[i])
    pyplot.show()
    

```

## 여성

```python
for i in range(itr):
    df_w.loc[np.power(df_w.sub(dfW.iloc[i], axis=1),2).sum(axis=1).sort_values().index[:5]].T.plot()
    pyplot.title(dfW.index[i])
    pyplot.show()
```

## 남성

```python
for i in range(itr):
    df_m.loc[np.power(df_m.sub(dfM.iloc[i], axis=1),2).sum(axis=1).sort_values().index[:5]].T.plot()
    pyplot.title(dfW.index[i])
    pyplot.show()
```

# 결론

* __`흑석동`__ 은 기본적으로 오래된 동네이지만 대학으로 인해 20대 비율이 좀 있는 동네들을 생각했다. 노량진, 광진구(건대,세종대),동대문구(경희대,외대,시립대) 등이 그래서 비슷하게 나왔다.
* __`고촌읍`__ 은 유소년층과 그 부모 세대가 많은 동네들을 생각했다. 학원가가 많은 장안동, 경기도 광주, 그리고 젊은 공무원들이 많은 세종시와 비슷하게 나왔다.
* __`영통1동`__ 은 대학 도시를 생각했지만 대학생들보다는 직장인이 많고, 경희대 학생들은 영통1동과 서천동에 분산되어 있기 때문에 대학생들 비중이 적었다.
